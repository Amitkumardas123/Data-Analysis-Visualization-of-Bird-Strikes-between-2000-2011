```

Feet above ground: The altitude (in feet) above ground level at which the bird strike occurred.

Number of people injured: The number of people who were injured as a result of the bird strike.

Is Aircraft Large?: A binary indicator (e.g., Yes/No) showing whether the aircraft involved is considered large (typically commercial jets).
```

# **Number of Bird strikes in yearly Basic.**

````
# Create the line plot
plt.figure(figsize=(10, 6))
plt.plot(yearly_bird_strikes.index, yearly_bird_strikes.values, marker='o')

# Add title and labels
plt.title('Yearly Bird Strikes', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Bird Strikes', fontsize=14)

# Annotate each point with its value
for x, y in zip(yearly_bird_strikes.index, yearly_bird_strikes.values):
    plt.text(x, y, f'{y}', fontsize=10, ha='right', va='bottom')

# Display the plot
plt.xticks(yearly_bird_strikes.index, rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjusts plot to ensure everything fits without overlapping
plt.show()
````

# **Altitude of aeroplanes at the time of strike**

````
# Group by 'Altitude bin' to count the occurrences
altitude_counts = Data['Altitude bin'].value_counts()
````

````
# Plotting the donut chart
plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(
    altitude_counts,
    labels=None,  # Labels will be customized below
    autopct='',   # Disable the default percentage display
    startangle=140,
    colors=plt.cm.Paired.colors,
    wedgeprops={'width': 0.4}  # This creates the hole in the middle (donut)
)

# Create custom labels with the number and percentage below the altitude bin
custom_labels = [f'{label}\n{count} ({round(100 * count / altitude_counts.sum(), 1)}%)'
                 for label, count in zip(altitude_counts.index, altitude_counts)]

# Reposition the labels
for i, text in enumerate(texts):
    text.set_text(custom_labels[i])
    text.set_size(12)  # Adjust label size if necessary

# Ensure the chart is a circle
plt.axis('equal')

# Adjust the title position to avoid overlapping
plt.title('Altitude of Aeroplanes at the Time of Bird Strike', pad=40)

# Apply tight layout to adjust everything nicely
plt.tight_layout()

# Show plot
plt.show()
````

***Majority of Bird Strikes at Lower Altitudes:*** A significant majority of bird strikes, accounting for 80.7% (19,977 instances), occur at altitudes below 1000 feet. This indicates that the risk of bird strikes is considerably higher during takeoff and landing phases of flight, when aircraft are at lower altitudes.

***Lower Incidence at Higher Altitudes: ***Only 19.3% (4,770 instances) of bird strikes happen at altitudes above 1000 feet. This suggests that while bird strikes can occur at higher altitudes, they are much less frequent compared to lower altitudes.

# **Professional Implications:**
***Aviation Safety Measures:*** The data underscores the importance of implementing robust bird strike prevention measures, particularly during the critical phases of takeoff and landing. Airports and airlines may need to focus their mitigation strategies on these lower altitude operations.

***Wildlife Management:*** Effective wildlife management practices around airports are crucial to reduce the risk of bird strikes. This could involve habitat modification, use of bird deterrent technologies, and regular monitoring of bird activity.

# **Average Altitude of the aeroplanes in different phases at the time of strike**

````
''' Filter out the unwanted ['Landing Roll', 'Parked', 'Take-off run', 'Taxi'] phases because
at that time plane is in ground'''
phases_to_remove = ['Landing Roll', 'Parked', 'Take-off run', 'Taxi']
filtered_df = Data[~Data['When: Phase of flight'].isin(phases_to_remove)]

# Calculate the average altitude for the remaining phases of flight
avg_altitude_per_phase = filtered_df.groupby('When: Phase of flight')['Feet above ground'].mean()
````

````
# Plotting the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(avg_altitude_per_phase.index, avg_altitude_per_phase, color=plt.cm.Paired.colors)

# Adding the average altitude above each bar
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{int(height)} ft',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom')

# Title and labels
plt.title('Average Altitude of Aeroplanes in Different Phases at the Time of Strike')
plt.xlabel('Phase of Flight')
plt.ylabel('Average Altitude (Feet)')

# Rotate x labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
````

# **Effect of Strike at Different Altitude**
